```python
# Exemplo cifra de Cesar

texto = 'Este e um texto cifrado'
chave = 1
modo = 'E'
CARACTERES = 'ABCDEFGHIJKLMNOPQRSTUVWXYZ'
convertido = ''
texto = texto.upper()
for caractere in texto:
    if caractere in CARACTERES:
        num = CARACTERES.find(caractere)
        if modo == 'E':
            num = num + chave
        if modo == 'D':
            num = num - chave
        if num >= len(CARACTERES):
            num = num - len(CARACTERES)
        if num < 0:
            num = num + len(CARACTERES)
        convertido = convertido + CARACTERES[num]
    else:
        convertido = convertido + caractere
if modo == 'E':
    print('O texto criptografado é ', convertido)
elif modo == 'D':
    print('O texto decriptado é ', convertido)
else:
    print('Opção inválida')
```

```python
# Exemplo Hash de Arquivos

import hashlib

def open_txt(file):
  with open(file) as f:
    return "".join(f.read())

file_1 = '/content/drive/MyDrive/Colab_Notebooks/Aula_6/a .txt'
file_2 = '/content/drive/MyDrive/Colab_Notebooks/Aula_6/c.txt'
text_1 = open_txt(file_1)
text_2 = open_txt(file_2)

def compare(text_1, text_2):
  hash_1 = hashlib.md5()
  hash_1.update(text_1.encode('utf-8'))
  hash_1 = hash_1.hexdigest()

  hash_2 = hashlib.md5()
  hash_2.update(text_2.encode('utf-8'))
  hash_2 = hash_2.hexdigest()

  return hash_1, hash_2;

hash = compare(text_1, text_2)
print(hash[0])
print(hash[1])
```

```python
# Atividade 2

import os
import hashlib

def open_txt(file):
    with open(file, 'r') as f:
        return f.read()

def get_md5(text):
    hash_obj = hashlib.md5(text.encode('utf-8'))
    return hash_obj.hexdigest()

pasta = '/content/drive/MyDrive/Colab_Notebooks/Aula_6/'

hashes = {}

for arquivo in os.listdir(pasta):
    if arquivo.endswith('.txt'):
        caminho_completo = os.path.join(pasta, arquivo)
        conteudo_arquivo = open_txt(caminho_completo)
        hash_md5 = get_md5(conteudo_arquivo)
        if hash_md5 in hashes:
            hashes[hash_md5].append(arquivo)
        else:
            hashes[hash_md5] = [arquivo]

for hash_md5, arquivos in hashes.items():
    if len(arquivos) > 1:
        print(f'Os arquivos {arquivos} são idênticos e têm o hash MD5: {hash_md5}')
```

O sistema de criptografia de chave pública conhecido como RSA usa duas chaves, uma privada e uma pública. A chave privada é usada para descriptografar mensagens e deve ser mantida em segredo pelo proprietário, enquanto a chave pública pode ser compartilhada abertamente para que outras pessoas a usem para criptografar mensagens que pertencem ao proprietário da chave. O princípio de segurança do RSA é baseado na dificuldade computacional de fatorar um número considerável em seus dois componentes primos originais, o problema da fatoração de inteiros. Este método é comumente usado para proteger dados em transferências seguras na web, como transações financeiras e mensagens confidenciais. Também pode ser usado para assinaturas digitais que garantem que os documentos são verdadeiros.
