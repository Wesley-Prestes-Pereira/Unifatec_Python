```python
# Exercicio 5

numero=int(input('Digite um número: '));
produto=1;
print('Números entre 1 e',numero,end=': ');
for contador in range(1, numero + 1):
  produto = produto * numero;
  if(contador<numero):
    print(contador, end=',');
  else:
    print(contador);
print('O produto de ',numero,' é ', produto);
```
