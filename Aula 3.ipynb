```python
# Exercício 3

vetor = [1, 2, 3, 4, 5]
i = 0
todos_positivos = True
while i < len(vetor) and todos_positivos:
    if vetor[i] <= 0:
        todos_positivos = False
    i += 1

if todos_positivos:
  print("Todos os elementos do vetor são positivos")
else:
  print("Todos os elementos do vetor são não são todos positivos")
```

```python
# Exercício 4

def calcular_media(vetor):
    return sum(vetor) / len(vetor)


vetor = [1, 2, 3, 4, 5]
media = calcular_media(vetor)
print("Média aritmética dos elementos do vetor:", media)
```

```python
# Exercício 5

def calcular_media(vetor):
    return sum(vetor) / len(vetor)


vetor = [1, 2, 3, 4, 5]
vetor.pop(4)
media_apos_remocao = calcular_media(vetor)
print("Média aritmética dos elementos do vetor:", media_apos_remocao)
```

```python
# Exercício  6

def verificar_strings_iguais(str1, str2):
    return str1 == str2

str1 = str(input('Digite uma palavra:'))
str2 = str(input('Digite outra palavra:'))
if(verificar_strings_iguais(str1, str2)):
  print("As palavras ",str1," e ",str2," são iguais")
else:
  print("As palavras ",str1," e ",str2," não são iguais")
```

```python
# Exercício  9

def horas(segundos):
    return segundos / 3600

# Exemplo de uso da função
segundos = int(input("Digite o tempo em segundos:"))
print(segundos,"segundos é equivalente a",horas(segundos),"horas")
```

```python
# Exercício  10
def sucessor_antecessor(numero):
    return numero + 1, numero - 1

# Exemplo de uso da função
numero = int(input("Digite um número: "))
sucessor, antecessor = sucessor_antecessor(numero)
print("Sucessor e antecessor de", numero,": ",sucessor,",",antecessor)
```
